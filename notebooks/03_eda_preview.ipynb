```python
features = {
    "has_5g": "5G",
    "has_nfc": "NFC",
    "has_ir_blaster": "IR Blaster",
    "fast_charging": "Fast Charging",
    "memory_card_supported": "Memory Card",
    "os_name": "Operating System",
}

fig = make_subplots(
    rows=2,
    cols=3,
    specs=[[{"type": "domain"}]*3, [{"type": "domain"}]*3],
    subplot_titles=list(features.values())
)

for idx, (col, title) in enumerate(features.items(), start=1):
    r = 1 if idx <= 3 else 2
    c = idx if idx <= 3 else idx - 3

    fig.add_trace(
        go.Pie(
            labels=df[col].value_counts().index,
            values=df[col].value_counts().values,
            showlegend=False,
            marker=dict(colors=px.colors.sequential.Purp_r),
            hole=0.3,
        ),
        row=r, col=c
    )

fig.update_layout(
    template="plotly_dark",
    title="Feature Availability",
    height=600,
    width=800,
)

# Adjust annotation lift
for ann in fig["layout"]["annotations"]:
    ann["y"] += 0.03

fig.show()
```

```python
temp_df = df["processor_brand"].value_counts().reset_index()

fig = px.bar(
    temp_df,
    x=temp_df["processor_brand"],
    y=temp_df["count"],
    title="Processor Brand Share",
    color="processor_brand",
    color_discrete_sequence=px.colors.sequential.Purp_r,
    template="plotly_dark",
    height=400,
    width=800,
)

fig.update_layout(showlegend=False)

fig.show()
```

```python
plots = {
    "battery_mah": "Battery Capacity Variation Across Smartphones",
    "charging_watt": "Charging Wattage Distribution"
}

colors = px.colors.sequential.Purp_r

fig = make_subplots(
    rows=1, cols=2,
    shared_xaxes=True,
    vertical_spacing=0.05,
    subplot_titles=list(plots.values())
)

for i, (col, title) in enumerate(plots.items(), start=1):
    fig.add_trace(
        go.Violin(
            y=df[col],
            name=title,
            marker=dict(color=colors[i * 2]),
            points="all"
        ),
        row=1, col=i
    )

fig.update_layout(
    template="plotly_dark",
    title="Battery Section Check",
    height=600,
    width=800,
    showlegend=False
)

fig.show()
```

# üîé Univariate Analysis ‚Äì Key Conclusions  
---

- **5G Support**
  - A large portion of the phones already support 5G, showing how quickly the market is shifting toward newer network technologies.
  - This also means 5G may influence price, since it is more common in mid-range and premium devices.

- **NFC Support**
  - NFC is far less common than 5G.
  - It usually appears in higher-priced or more feature-rich phones.
  - Its limited availability makes it a good indicator of whether a device belongs to a mid-range or premium category.

- **IR Blaster**
  - Only a small share of devices offer an IR blaster.
  - This feature is brand-dependent (mainly Xiaomi).
  - It doesn‚Äôt strongly relate to pricing across the dataset.

- **Fast Charging**
  - Fast charging is almost universally available, even on budget devices.
  - This shows it has become a standard feature rather than a premium differentiator.

- **Operating System**
  - The dataset is dominated by Android, with very few iOS devices.
  - This means the price distribution is shaped primarily by Android‚Äôs mid-range and budget ecosystem.

- **Processor Brand Share**
  - MediaTek is the most widely used processor brand in the dataset, appearing in a large number of devices.
  - Snapdragon holds the second position, commonly seen in mid-range and upper-mid devices.
  - Unisoc and Exynos appear in smaller proportions, mostly in budget and mid-tier phones.
  - Apple and Google Tensor have very few entries, which is expected since iPhones and Pixels make up a small portion of the dataset.
  - Overall, the processor distribution confirms that the dataset is dominated by **Android mid-range hardware**, with only a limited presence of premium flagship chipsets.

- **Smartphones Brand Share**
  - The dataset is dominated by well-known mid-range brands like **Realme, Samsung, Vivo, Oppo, and Xiaomi**, which make up the majority of the entries.
  - Brands such as **Motorola, Lava, Poco, Tecno, iQOO, and Infinix** appear moderately, reflecting their strong presence in the budget and mid-budget segment.
  - Premium or niche brands like **Apple, OnePlus, Nothing, and Google** appear far less frequently.
  - The distribution clearly shows that the dataset is heavily oriented toward **budget and mid-range Android smartphones**, with only a small representation from high-end flagship brands.

- **Clock Speed (GHz)**
  - Most smartphones in the dataset operate within a fairly narrow clock-speed range, typically around **1.8 GHz to 2.4 GHz**.
  - This distribution is expected because modern mid-range processors usually cluster around similar base frequencies.
  - Very high or very low clock speeds are rare, indicating that outliers are minimal for this feature.
  - Overall, clock speed alone does not strongly differentiate devices but works best when combined with other factors like core count and processor brand.

- **Number of Cores Used**
  - The majority of smartphones in the dataset come with **8-core processors**, which aligns with most modern Android devices.
  - The next most common are **6-core processors**, used mainly in **Apple A-series chips** (2 performance + 4 efficiency cores) and a few mid-range Android CPUs.
  - **4-core processors** appear in older or budget smartphones that are still part of the dataset.
  - A very small number of devices use **2-core** or even **1-core** CPUs, indicating extremely low-end or outdated models.
  - **10-core processors** show up only in rare cases (old MediaTek Helio X-series), making them clear outliers.
  - Since 8-core and 6-core devices dominate, core count alone does not fully explain performance ‚Äî you need to combine it with **processor brand** and **clock speed** for meaningful insights.

- **Display Size (inches)**
  - Most smartphones in the dataset fall in the **6.4 to 6.8 inch** range, which represents today‚Äôs standard display size for mid-range devices.
  - A small number of phones have **compact screens around 5.4‚Äì5.8 inches**, indicating older or rare small-form-factor models.
  - A few devices extend beyond **7 inches**, entering the phablet/tablet-like category (e.g., foldables or extra-large displays).
  - The overall distribution is heavily concentrated toward larger screens, showing that modern smartphones increasingly favor big displays.
  - Display size on its own doesn‚Äôt strongly separate price categories, but when combined with **resolution** and **refresh rate**, it becomes more meaningful for differentiating budget vs premium models.

- **Battery Capacity (mAh)**
  - Most smartphones in the dataset have a battery capacity between **4500 mAh and 6000 mAh**, with **5000 mAh** being the most common.
  - A few devices include larger batteries in the **6500‚Äì7000 mAh** range, usually meant for heavy-use or rugged models.
  - There is one **extreme outlier at 21,200 mAh**, which is far beyond normal smartphone capacity.  
    This value will distort analysis and should be removed before model training.
  - A small number of older or budget phones fall below **3000 mAh**, forming a low-end outlier group.
  - Overall, the battery distribution is **right-skewed**, with most phones clustering around typical modern capacities.

- **Charging Wattage**
  - Most phones offer **18W to 33W** fast charging, which has become the industry standard.
  - Higher wattage options like **44W, 50W, 67W, and 120W** appear in mid-range and premium phones.
  - A few models reach extremely high charging speeds, but these are rare compared to the rest of the dataset.
  - Charging wattage shows a wider spread than battery size, but still follows a clear pattern where higher wattage is found in higher-priced devices.

- **Front Camera Count**
  - Most smartphones in the dataset come with a **single front camera**, which is the standard setup across the industry.
  - A smaller number of devices use **dual front cameras**, usually found in selfie-focused or premium models.
  - Front camera counts are generally simple and do not vary much, so they contribute only limited differentiation among devices.

- **Rear Camera Count**
  - The majority of smartphones feature **triple-camera or quad-camera** setups on the rear.
  - Dual-camera phones appear in moderate numbers, typically in older or budget models.
  - Single-camera devices are very rare and mostly belong to older generations.
  - Overall, rear camera count shows more variety than the front camera count and can help separate budget phones (2 cameras) from mid-range and premium phones (3‚Äì4 cameras).

- **Front Camera Megapixel Distribution**
  - Most smartphones have front cameras in the **8 MP to 16 MP** range, which is the standard for mid-range devices.
  - A few models offer **20 MP or higher** front cameras, usually targeting users who prioritize selfies.
  - Extremely low front MP values appear in a small number of older or entry-level phones.
  - Overall, the distribution is fairly concentrated, meaning front MP does not vary wildly across phones.

- **Rear Camera Megapixel Distribution**
  - Rear camera megapixels show a **much wider spread** compared to front cameras.
  - Common values include **12 MP, 48 MP, 50 MP, and 64 MP**, which are popular across many brands.
  - Some models reach very high MP counts like **108 MP** or **200 MP**, which are typical of premium or camera-centric phones.
  - The distribution has clear peaks at marketing-driven megapixel values (e.g., ‚Äú48MP‚Äù, ‚Äú64MP‚Äù, ‚Äú108MP‚Äù).

- **Overall Comparison(Front Camera Megapixel Distribution and Rear Camera Megapixel Distribution)**
  - Rear cameras show **far more variation** than front cameras, both in range and in peak values.
  - Higher MP counts on the rear side often indicate **better camera hardware** and may correlate with higher prices.
  - Front cameras are more standardized, and megapixel differences are less dramatic, making them a weaker indicator of pricing.

- **Memory Card Support**
  - A noticeable portion of smartphones **do not support memory cards**, especially newer mid-range and premium models.
  - Phones that do support memory expansion are mostly from budget and mid-budget segments.
  - The absence of a memory card slot often indicates a more modern or premium device design.
  - This feature may help distinguish lower-priced devices from higher-priced ones.

- **Memory Card Type**
  - Among the phones that support external storage, **microSD** is the dominant and nearly universal card type.
  - Very few models, if any, use alternative card formats.
  - Since type only matters when support is available, this feature adds limited additional value compared to the support flag itself.

- **Overall Observation(Memory Card Support and Memory Card Type)**
  - Memory card support is becoming less common in newer smartphones.
  - Devices without memory card support tend to fall into the **mid-range and premium categories**, which may impact pricing.
  - Memory card type has little variation and should be used only alongside the support indicator.
