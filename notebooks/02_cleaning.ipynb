```python
def get_mp_list(text):
    """Extract all MP numbers"""
    if pd.isna(text):
        return []
    return [float(x) for x in re.findall(r'(\d+(?:\.\d+)?)\s*MP', text)]
```

```python
def primary_rear(text):
    if pd.isna(text):
        return np.nan
    rear_part = text.split("&")[0]
    mps = get_mp_list(rear_part)
    return max(mps) if mps else np.nan
```

```python
def primary_front(text):
    if pd.isna(text):
        return np.nan
    parts = text.split("&")
    if len(parts) > 1:
        front_part = parts[1]
    else:
        return np.nan
    mps = get_mp_list(front_part)
    return max(mps) if mps else np.nan
```

```python
def dual_rear(text):
    if pd.isna(text):
        return False
    rear_part = text.split("&")[0]
    mps = get_mp_list(rear_part)
    return len(mps) >= 2
```

```python
def dual_front(text):
    if pd.isna(text):
        return False
    parts = text.split("&")
    if len(parts) > 1:
        front_part = parts[1]
    else:
        return False
    mps = get_mp_list(front_part)
    return len(mps) >= 2
```

```python
primary_rear = df1["camera"].apply(primary_rear)
primary_front = df1["camera"].apply(primary_front)
has_dual_rear = df1["camera"].apply(dual_rear)
has_dual_front = df["camera"].apply(dual_front)
```

```python
def parse_supported(x):
    if pd.isna(x):
        return np.nan
    text = str(x).lower()
    
    if "not supported" in text:
        return False
    if "memory card" in text:
        return True
    return np.nan
```

```python
memory_card_supported = df1["memory_card"].apply(parse_supported)
```

```python
def parse_type(x):
    if pd.isna(x):
        return np.nan
    text = str(x).lower()
    
    if "hybrid" in text:
        return "hybrid"
    if "supported" in text:
        return "dedicated"
    return np.nan
```

```python
memory_card_type = df1["memory_card"].apply(parse_type)
```

```python
def parse_size(x):
    if pd.isna(x):
        return np.nan
    text = str(x).lower()

    match = re.search(r'upto\s*([\d\.]+)\s*(tb|gb)', text)
    if match:
        val = float(match.group(1))
        unit = match.group(2)

        if unit == "tb":
            return int(val * 1024)
        else:
            return int(val)
    return np.nan
```

```python
memory_card_max_gb = df1["memory_card"].apply(parse_size)
```
